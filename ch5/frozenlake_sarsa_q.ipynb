### SARSA

```python
# FrozenLake-v1
# Q(s, a)를 key '(s, a)'로 가진 Dictionary 데이터로 정의하고 초기치를 Uniform (0, 1)로 부여하고 있음.

import gym
env = gym.make('FrozenLake-v1')

Q = {}

for s in range(env.observation_space.n):
    for a in range(env.action_space.n):
        Q[(s, a)] = random.uniform(0, 1)
```

```python
# Epsilon-Greedy Policy를 정의함.

def epsilon_greedy(state, epsilon):
    if random.uniform(0, 1) < epsilon:
        return env.action_space.sample()
    else:
        return max(list(range(env.action_space.n)), key=lambda x: Q[(state, x)])
```

```python
# Q(s, a)를 Update하기 위해 학습률 alpha = 0.1, discount factor gamma = 0.9, Epsilon-Greedy Policy를 위해 epsilon = 0.2로 값을 부여함.ㄹ
# Q(s_t+1, a_t+1) --- Q(s_, a_)에서 action a_의 선택은 epsilon-greedy policy로 결정하고 있음.

alpha = 0.1
gamma = 0.9
epsilon = 0.2

num_episodes = 200000
timesteps = 1000

for i in range(num_episodes):
    s = env.reset()
    a = epsilon_greedy(s, epsilon)
    
    for t in range(timesteps):
        s_, r, done, _ = env.step(a)
        a_ = epsilon_greedy(s_, epsilon)
        
        Q[(s, a)] = Q[(s, a)] + alpha * (r + gamma * Q[(s_, a_)] - Q[(s, a)])
        s = s_
        a = a_
        
        if done:
            break 
```

```python
# 최종 추정된 Q(s, a)를 출력

# State 0에서 최적 Policy는 2 (action = 2, right) - Value가 0.451180으로 가장 높으므로.
# State 1에서 최적 Policy는 2 (action = 2, right) - Value가 0.507997로 가장 높으므로.

import pandas as pd
df = pd.DataFrame(list(Q.items()), columns=['state_action_pair', 'value'])
df.head(20)
```

### Q-Learning
- Q-Learning은 Q(s_t+1, a_t+1)을 계산하기 위해 선택하는 a_t+1를 argmax a' Q(s_t+1, a')으로 바꾼 것을 제외하고는 SARSA 프로그램과 동일함.

```python
import gym
import random
import numpy as np
```

```python
env = gym.make('FrozenLake-v1')

Q = {}

for s in range(env.observation_space.n):
    for a in range(env.action_space.n):
        Q[(s, a)] = random.uniform(0, 1)
```

```python
def epsilon_greedy(state, epsilon):
    if random.uniform(0, 1) < epsilon:
        return env.action_space.sample()
    else:
        return max(list(range(env.action_space.n)), key=lambda x: Q[(state, x)])
```

```python
alpha = 0.1
gamma = 0.9
epsilon = 0.2

num_episodes = 200000
timesteps = 1000

for i in range(num_episodes):
    s = env.reset()
    
    for t in range(timesteps):
        a = epsilon_greedy(s, epsilon)
        s_, r, done, _ = env.step(a)
        a_ = np.argmax([Q[s_, a] for a in range(env.action_space.n)])
        
        Q[(s, a)] = Q[(s, a)] + alpha * (r + gamma * Q[(s_, a_)] -  Q[(s, a)])
        s = s_
        
        if done:
            break
```

```python
# Q-Learning에 의해 학습된 Q(s, a)를 보여주고 있음.

# SARSA의 Q(s, a)에 비교하여 Q-Learning의 Q(s, a) 값이 전반적으로 큰 것을 볼 수 있음.
# max a' Q(s_t+1, a')에 의해 과대추정되었기 때문임.

import pandas as pd

df = pd.DataFrame(list(Q.items()), columns=['state_action_pair', 'value'])
print(df.head(20))
```
